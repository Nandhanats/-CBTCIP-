```python
import pandas as pd
```

```python
data = pd.read_csv('/content/Unemployment in India - Unemployment in India.csv')
```

```python
print("First few rows of the dataset:")
print(data.head())
```

```python
print("\nLast few rows of the dataset:")
print(data.tail())
```

```python
print(data.shape)
```

```python
print(data.columns)
```

```python
print(data.describe())
```

```python
print(data.isnull().sum())
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
import matplotlib.pyplot as plt
```

 Unemployment Rate

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Estimated Unemployment Rate (%)'], marker='o', linestyle='-')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Employment Rate

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], 100 - data['Estimated Unemployment Rate (%)'], marker='o', linestyle='-')
plt.title('Employment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Employment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Labor Participation Rate

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Estimated Labour Participation Rate (%)'], marker='o', linestyle='-')
plt.title('Labor Participation Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Labor Participation Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Analysis by Region

```python
region_data = data.groupby('Region').mean()
region_data.plot(kind='bar', y=['Estimated Unemployment Rate (%)', 'Estimated Labour Participation Rate (%)'], figsize=(10, 6))
plt.title('Average Unemployment and Labor Participation Rate by Region')
plt.ylabel('Percentage')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Analysis by Frequency

```python
frequency_data = data.groupby('Frequency').mean()
frequency_data.plot(kind='bar', y=['Estimated Unemployment Rate (%)', 'Estimated Labour Participation Rate (%)'], figsize=(10, 6))
plt.title('Average Unemployment and Labor Participation Rate by Frequency')
plt.ylabel('Percentage')
plt.xlabel('Frequency')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Filtering Data by Region

```python
region_data = data[data['Region'] == 'Region_Name']
print(region_data.head())
```

Sorting Data by a Column

```python
sorted_data = data.sort_values(by='Estimated Unemployment Rate (%)', ascending=False)
print(sorted_data.head())
```

```python
mean_unemployment_by_region = data.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
print(mean_unemployment_by_region)
```

```python
correlation_matrix = data.corr()
print(correlation_matrix)
```

```python
import seaborn as sns
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
